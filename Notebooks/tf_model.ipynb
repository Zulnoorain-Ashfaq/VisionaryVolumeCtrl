```python
import tensorflow as tf
import cv2
from keras import layers
```

```python
model = tf.keras.Sequential([
    layers.Conv2D(64,3, input_shape=(256,256,1)),
    layers.Conv2D(64,3),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.AveragePooling2D(),
    layers.Dropout(.5),

    layers.Conv2D(128,3),
    layers.Conv2D(128,3),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.AveragePooling2D(),
    layers.Dropout(.5),

    layers.Conv2D(256,3),
    layers.Conv2D(256,3),
    layers.BatchNormalization(),
    layers.ReLU(),
    layers.AveragePooling2D(),

    layers.Flatten(),

    layers.Dense(1000, activation='relu'),
    layers.Dropout(.5),
    layers.Dense(1,activation='sigmoid')
])
model.compile('adam','binary_crossentropy',['accuracy'])
```

```python
import time
import cv2
from ctypes import cast, POINTER
from comtypes import CLSCTX_ALL
from pycaw.pycaw import AudioUtilities, IAudioEndpointVolume

classs = ['Focused', 'distracted']
cap = cv2.VideoCapture(0)
```

```python
tolerance = 4
current_tol = 0
devices = AudioUtilities.GetSpeakers()
interface = devices.Activate(
   IAudioEndpointVolume._iid_, CLSCTX_ALL, None)
volume = cast(interface, POINTER(IAudioEndpointVolume))
played = False
while cap.isOpened():
    ret, image = cap.read()
    print(image.shape)

    face_detector = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_alt2.xml")

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Detect faces in the image
    faces = face_detector.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
    # Draw rectangles around the faces
    out = 1
    for (x, y, w, h) in faces:
        face = image[y:y+h, x:x+w]
        face = cv2.cvtColor(face,cv2.COLOR_RGB2GRAY)
        face = cv2.resize(face,(256,256))
        cv2.imshow('face',face)
        cv2.waitKey(1)
        out = round(model([face.reshape(256,256,1)])[0][0].numpy())
        # out = round(model.predict(face.reshape(1,256,256,1),verbose=False)[0][0])

    # for decreasing volume
    current_volume = volume.GetMasterVolumeLevel()
    if out == 0 and current_volume!=0.0:
        volume.SetMasterVolumeLevel(0.0, None)
        current_tol = 0
    elif out == 1 and current_volume!=-63:
        current_tol += 1

    # for pausing
    # if out == 0 and not played:
    #     pyautogui.press('playpause')
    #     current_tol = 0
    #     played = True
    # elif out == 1:
    #     current_tol += 1

    if current_tol == tolerance:
        # pyautogui.press('playpause')
        # played = False
        volume.SetMasterVolumeLevel(-63, None)
    time.sleep(.1)
```

```python
cap.release()
cv2.destroyAllWindows()
```
